## Visualization and Pandas

* Visualization is an essential tool to make data easier to understand.
* One picture says more than 1000 words.
* Pythonâ€™s data analysis library, Pandas, provides several different options for visualizing your data with `.plot()`.
* Pandas allows you to create basic charts to gain valuable insights from your data.
* Pandas collaborates with the Matplotlib library and enables the rapid generation of graphs.

## Examples of graphs in Pandas

```python
# data loading and preparation

import pandas as pd

df_countries = pd.read_csv("data/countries.csv")
df_countries['Population'] /= 1000000
df_countries.head()
```

##### Graph showing the proportion of the population from each region to the total population.

```python
import matplotlib.pyplot as plt

df_countries.groupby('Region')['Population'].sum().plot(kind='pie', label='', figsize=(7,7))
plt.show()
```

##### Populations from each region in the bar chart.

```python
import matplotlib.pyplot as plt

df_countries.groupby('Region')['Population'].sum().plot(kind='bar')
plt.show()
```

##### Graph showing the relationship between the area of the country and the size of the polulation.

```python
import matplotlib.pyplot as plt

df_countries.plot(x="Area sq. mi.", y="Population", kind="scatter")
plt.show()
```

##### Graph showing the number of Lego sets introduced by year.

```python
df = pd.read_csv('data/sets.csv')
df.groupby('year')['name'].count().plot()
plt.show()
```
