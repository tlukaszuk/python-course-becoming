## Multiple graphs on one "canvas"

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 11)
y = x ** 2

plt.subplot(1,2,1) # one row, two columns, graph one
plt.plot(x, y, 'b--')
plt.subplot(1,2,2) # one row, two columns, graph two
plt.plot(y, x, 'r*-')
plt.show()
```

## The main idea in the object-oriented approach in Matplotlib

* Creating objects and then calling methods on these objects or referring to the attributes of these objects.
* This approach already benefits from having more graphs on a single 'canvas'.

## Graph construction in Matplotlib

<img src="img\matplotlib_graph_structure.png" style="width:400px">

* Figure - an object that can be compared to the 'canvas' on which graphs are painted,
* Axes - the space of the chart, where we have axes, chart, title, labels, etc..,
* Axis - otherwise the axes of the graph, for 2D (x,y) for 3D (x,y,z).

## Creating a graph in an object-oriented approach

```python
fig = plt.figure() # we create a Figure object, which is our "canvas".

# we add a graph, in square brackets we have the position of the graph given
# 0.1 start from the left, 0.1 start from the bottom, 0.8 chart width, 0.8 chart height
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # values in the range (0 - 1)

# add data to chart, set labels and title
axes.plot(x, y, 'b--')
axes.set_xlabel('X axis name')
axes.set_ylabel('Y axis name')
axes.set_title('Here we have the title')
plt.show()
```

## Greater control over the graph

* E.g. two graphs on one 'canvas', a larger and a smaller one.

```python
fig = plt.figure() # we create a Figure object, i.e. our "canvas"

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # larger graph
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # smaller graph

# bigger graph
axes1.plot(x, y, 'b--')
axes1.set_xlabel('X-axis of larger graph')
axes1.set_ylabel('Y-axis of bigger graph')
axes1.set_title('Title of larger chart')

# The smaller graph
axes2.plot(y, x, 'r*-')
axes2.set_xlabel('X-axis of smaller graph')
axes2.set_ylabel('Y-axis of smaller graph')
axes2.set_title('Title of smaller graph')
plt.show()
```

## Matplotlib chart manager

* `plt.subplots()` - we pass the information that Matplotlib has to take care of the layout of the charts, it works a bit like an automatic chart manager.

```python
fig, axes = plt.subplots()

# we use the axes object to add data to the plot
axes.plot(x, y, 'r')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('title')
plt.show()
```

* However, we can do more than that. We can specify the number of rows and columns when calling the `plt.subplots()` method.

```python
# Empty 'canvas' with two plots waiting for data
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.show()
```

* The `axes` object is actually a list of graphs, in our case two empty graphs.

```python
for ax in axes:
    ax.plot(x, x, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# automatically adjust the position of the two graphs so that they do not overlap
fig.tight_layout()

# display the 'canvas', i.e. the Figure object  
fig
```

## Canvas size and DPI

* Methods specification: `plt.figure(figsize, dpi)`, `plt.subplots(figsize, dpi)`
* `figsize` - a tuple in which we specify width and height in inches
* `dpi` - dots-per-inch or number of pixels per inch


```python
fig, axes = plt.subplots(figsize=(5,2),dpi=100)
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
plt.show()
```

## Titles, axis names and legend

* `ax.set_title("title")` - title the graph
* `ax.set_xlabel("x")`, `ax.set_ylabel("y")` - naming of axes
* `label = "label for data"` - parameter used to specify labels in the legend

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(X, S, label="sinus")
ax.plot(X, C, label="cosinus")
ax.legend()
plt.show()
```

#### Legend position:

* `ax.legend(loc=0)` - default most optimal position
* `ax.legend(loc=1)` - upper right corner
* `ax.legend(loc=2)` - upper left corner
* `ax.legend(loc=3)` - bottom left corner
* `ax.legend(loc=4)` - bottom right corner

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(X, S, label="sinus")
ax.plot(X, C, label="cosinus")
ax.legend(loc=4)
plt.show()
```

## Colours, styles and line thickness

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(X, S, 'b--', label='sinus') # blue colour and dashed line
ax.plot(X, C, 'g:', label='cosine') # green colour and dotted line
ax.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 10, endpoint=True)
fig, ax = plt.subplots(figsize=(10,5))

# different line thicknesses
ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+4, color="red", linewidth=2.00)

# line styles possible options: '-', '-.', ':'
ax.plot(x, x+5, color='green', lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# possible marker styles: '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color='blue', lw=1, ls='-', marker='+')
ax.plot(x, x+10, color='blue', lw=1, ls='--', marker='o')
ax.plot(x, x+11, color='blue', lw=1, ls='-', marker='s')

# size and colours of markers
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
plt.show()
```

## Boundaries of the graph

* Definition of which part of the graph (function) is to be displayed.

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].plot(C)
axes[0].plot(S)
axes[0].set_title("default chart boundaries")
axes[1].plot(C)
axes[1].plot(S)
axes[1].axis('tight')
axes[1].set_title('tight_automatically')
axes[2].plot(C)
axes[2].plot(S)
axes[2].set_ylim([0,1])
axes[2].set_xlim([50, 200])
axes[2].set_title("manual_limit")
plt.show()
```
